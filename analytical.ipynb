
$\varphi=2 \theta+\sin 2 \theta =\quad\frac{2 x y}{x^2+y^2}+2 \tan ^{-1}\left(\frac{y}{x}\right)$
$$
\begin{aligned}
\therefore \quad V_\gamma & =\frac{1}{r} \frac{\partial \varphi}{\partial \theta}=\frac{1}{r} \cdot(2+2 \cos 2 \theta) \\
\nu_\theta & =0 \\
\because \quad V_x & =V r \cos \theta-\nu \theta \sin \theta \\
V_y & =V r \sin \theta+V \theta \cos \theta \\
V & =\sqrt{x^2+y^2}, \cos x=\frac{x}{\sqrt{x^2+y^2}} \quad, \sin x=\frac{y}{\sqrt{x^2+y^2}} \\
\therefore V_x & =\frac{1}{r}(2+2 \cos 2 \theta) \cos \theta \\
& =\frac{2}{\sqrt{x^2+y^2}} \cdot \frac{x}{\sqrt{x^2+y^2}}\left(1+\frac{x^2-y^2}{x^2+y^2}\right) \\
& =\frac{2 x}{x^2+y^2}\left(\frac{2 x^2}{x^2+y^2}\right) \\
& =\frac{4 x^3}{\left(x^2+y^2\right)^2}
\end{aligned}
$$
$$
\begin{aligned}
V_y & =\frac{1}{r} \cdot(2+2 \cos 2 \theta) \sin \theta \\
& =\frac{2}{\sqrt{x^2+y^2}} \cdot \frac{y}{\sqrt{x^2+y^2}}\left(1+\frac{x^2-y^2}{x^2+y^2}\right) \\
& =\frac{2 y}{x^2+y^2}\left(\frac{2 x^2}{x^2+y^2}\right) \\
& =\frac{4 x^2 y}{\left(x^2+y^2\right)^2} \\
\because V_x & =\frac{\partial \varphi}{\partial y}, \quad V_y=-\frac{\partial \varphi}{\partial x}
\end{aligned}
$$

The stress: $\zeta=\frac{\partial v_y}{\partial x}=\frac{\partial}{\partial x}\left[\frac{4 x^2 y}{\left(x^2+y^2\right)^2}\right] =\frac{8 y\left(x y^2-x^3\right)}{\left(x^2+y^2\right)^3}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Stream function plot
x = np.linspace(0, 1, 100)
y = np.linspace(-0.5,0.5, 100)
X, Y = np.meshgrid(x, y) 

lamda = 2/10
c = lamda**2/np.pi

Psi = (c/2) * (X * Y / (X**2 + Y**2) + np.arctan2(Y, X))

# Plot the streamlines
plt.figure(figsize=(8, 8))
plt.streamplot(X, Y, np.ones_like(Psi), Psi, color='k')
plt.contour(X, Y, Psi, colors='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stream Function Plot')
plt.axis('equal')
plt.show()
```

```python
#shear stress plot and Vx & Vy plot
x = np.linspace(0, 1, 100)
y = np.linspace(-0.5, 0.5, 100)
X, Y = np.meshgrid(x, y)

R_squared = X**2 + Y**2

Vx = 4 * X**3 / R_squared**2
Vy = 4 * X**2 * Y / R_squared**2

shear_stress = 8 * X * Y * (-X**2 + Y**2) / R_squared**3

plt.figure(figsize=(10, 8))

contour1 = plt.contourf(X, Y, shear_stress, levels=50, cmap='plasma') 
plt.colorbar(contour1)
plt.title('shear stress plot')

plt.figure(figsize=(10, 8))
contour = plt.contourf(X, Y,Vx, Vy, levels=100, cmap='turbo') 
plt.colorbar(contour) 

plt.title('Vx AND Vy plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)


plt.show()

print(shear_stress)
print(Vx)
print(Vy)
```
